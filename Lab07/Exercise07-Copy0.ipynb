### Estimating parameters of a logistic model

Gelman's book has an example where the dose of a drug may be affected to the number of rat deaths in an experiment.

| Dose (log g/ml) | # Rats | # Deaths |
|-----------------|--------|----------|
| -0.896          | 5      | 0        |
| -0.296          | 5      | 1        |
| -0.053          | 5      | 3        |
| 0.727           | 5      | 5        |

We will model the number of deaths as a random sample from a binomial distribution, where $n$ is the number of rats and $p$ the probabbility of a rat dying. We are given $n = 5$, but we believve that $p$ may be related to the drug dose $x$. As $x$ increases the number of rats dying seems to increase, and since $p$ is a probability, we use the following model:

$$
y \sim \text{Bin}(n, p) \\
\text{logit}(p) = \alpha + \beta x \\
\alpha \sim \mathcal{N}(0, 5) \\
\beta \sim \mathcal{N}(0, 10)
$$

where we set vague priors for $\alpha$ and $\beta$, the parameters for the logistic model.

```python
logit_code = """
data {
    int<lower=1> d;
    int<lower=0> n;
    int<lower=0> y[d];
    real x[d];
}
transformed data {}
parameters {
    real alpha;
    real beta;
}
transformed parameters {}
model {
    alpha ~ normal(0, 5);
    beta ~ normal(0, 10);
    for (i in 1:d){
        y[i] ~ binomial(n, inv_logit(alpha*x[i] + beta));
    }
}
generated quantities {}
"""

d = 4 # dimensions of x and y
n = 5 # number of rats in each category
# n = 5 * np.ones(4)
x = np.array([-0.896,-0.296,-0.053,0.727])
y = np.array([0,1,3,5])

logit_dat = {'d':d, 'n': n, 'x': x, 'y': y}


fit = pystan.stan(model_code=logit_code, data=logit_dat, iter=10000, warmup=5000, thin=2, chains=1)
# warmup = burin, thin = keep each ?th sample
```

### Using a hierarchical model

This uses the Gelman radon data set and is based on this [IPython notebook](http://twiecki.github.io/blog/2014/03/17/bayesian-glms-3/). Radon levels were measured in houses from all counties in several states. Here we want to know if the preence of a basement affects the level of radon, and if this is affected by which county the house is located in. 

The data set provided is just for the state of Minnesota, which has 85 counties with 2 to 116 measurements per county. We only need 3 columns for this example `county`, `log_radon`, `floor`, where `floor=0` indicates that there is a basement.

We will perfrom simple linear regression on log_radon as a function of county and floor.


```python
radon = pd.read_csv('radon.csv')[['county', 'floor', 'log_radon']]
radon.head()
```

```python
print min(radon['log_radon']), max(radon['log_radon']), np.mean(radon['log_radon']), np.median(radon['log_radon'])
print pd.unique(radon['floor'])
```

```python
# county = pd.Categorical(radon['county']).
# type(county.equals)
county = pd.Categorical(radon['county'])
county = county.labels + 1   # make it start from 1
np.max(county) # 85 (the number of unique counties)
```

```python
# NEED TO MAKE A HIERARCHICAL MODEL!
# y_ij = alpha_i * county_ij + beta_i * floor_ij + gamma + epsilon_ij (error term)
# n: the number of rows of the dataset (total number of observations)

# Simple regression model (not used)
# log_radon = alpha1*county + alpha2*floor + alpha3 + epsilon (error term)
# How to define categorical variables in pystan? Using "int" is fine.
# Should an interaction term, county*floor be included? No, county and floor are conditionally independent.
radon_code = """
data {
    int<lower=0> n; 
    int num_county;
    int<lower=0> county[n];
    real<lower=0> xfloor[n];
    real log_radon[n];    
}
transformed data{}
parameters {
    vector[num_county] a;
    vector[num_county] b;
    real mu_a;
    real mu_b;
    real<lower=0> sig_a;
    real<lower=0> sig_b;
    real<lower=0> sigma;
}
transformed parameters {
    real mu[n];
    for (i in 1:n) {
        mu[i] <- a[county[i]]*xfloor[i] + b[county[i]];
    }
}
model{
    a ~ normal(mu_a, sig_a);
    b ~ normal(mu_b, sig_b);
    mu_a ~ normal(0,100);
    mu_b ~ normal(0,100);
    sig_a ~ uniform(0,100);
    sig_b ~ uniform(0,100);
    sigma ~ gamma(2,2);
    log_radon ~ normal(mu,sigma);
}
generated quantities {}
"""

n = 919
num_county = 85

radon_dat = {'n':n, 'num_county':num_county, 'county':county, 'xfloor':radon['floor'], 'log_radon':radon['log_radon']}

fit2 = pystan.stan(model_code=radon_code, data=radon_dat, iter=10000, warmup=5000, thin=2, chains=1)
```

**Exercise 3 (10 points)**: Implement the leap-frog method and simulate the spring dynamics following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the spring trajectory plots.

```python
# Actually Example 1 in the URL
delta = 0.1
L = 70
def K(p):
    """Kinetic energy function"""
    return 0.5*(p**2)
def U(x):
    """Potential energy function for spring (K=1)"""
    return 0.5*x**2
def dU(x):
    """Gradient of potential energy"""
    return x

# Initial conditions
x0 = -4 # position
p0 = 1 # momentum

# Simulate Hamiltonian dynamics with leapfrog method
pStep = np.zeros(L)
xStep = np.zeros(L)

# First half step for momentum
pStep[0] = p0 - delta/2*dU(x0)
# First full step for position/sample
xStep[0] = x0 + delta*pStep[0]

# Full steps
for i in range(1,L-1):
    # update momentum
    pStep[i] = pStep[i-1] - delta*dU(xStep[i-1])
    # update position
    xStep[i] = xStep[i-1] + delta*pStep[i]

# Last half step for momentum
pStep[L-1] = pStep[L-2] - delta/2*dU(xStep[L-2])
# Update the position for the last time
xStep[L-1] = xStep[L-2] + delta*pStep[L-1]
```

**Exercise 4 (20 points)**: Now implement the HMC following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the bivariate normal MCMC plots. (For another simple explanation of HMC, see Chapter 14 of "Doing Bayesian Data Analysis" by Jouh Kruschke.) 

```python
# Actually Example 2 in the URL
np.random.seed(12345) # for this whole session/notebook (generates a sequence of random numbers with the seed)

# step size
delta = 0.3
nSamples = 1000
L = 20

def U(x):
    """Potential energy function"""
    return np.dot(np.dot(np.transpose(x),np.linalg.inv(np.array([[1,.8],[.8,1]]))),x)
def dU(x):
    """Gradient of potential energy"""
    return np.dot(np.transpose(x),np.linalg.inv(np.array([[1,.8],[.8,1]])))
def K(p):
    """Kinetic energy function"""
    return np.sum(p**2)*0.5  # equivalent to sum(t(p)%*%p)/2

# initial state
x = np.zeros((2,nSamples))
x0 = np.array([0,6])
x[:,0] = x0

t = 0

while t < (nSamples-1):
    # sample random momentum
    # np.random.seed(t+1) # NOT needed!
    p0 = np.random.random((2,1))
    
    # simulate Hamiltonian dynamics
    # first 1/2 step of momentum
    pStar = p0 - delta*0.5*dU(x[:,t]).reshape(2,1)
    
    # first full step for position/sample
    xStar = x[:,t].reshape(2,1) + delta*pStar
    
    # full steps
    for j in range(L-1):
        # momentum
        pStar = pStar - delta*dU(xStar).reshape(2,1)
        
        # position/sample
        xStar = xStar + (delta*pStar).reshape(2,1)
        
    # last half step
    pStar = pStar - (delta/2*dU(xStar)).reshape(2,1)
    
    # Evaluate energies at start and end of trajectory
    U0 = U(x[:,t])
    UStar = U(xStar)
    
    K0 = K(p0)
    KStar = K(pStar)
    
    # Acceptance/Rejection criterion
    alpha = min(1.0,np.exp((U0+K0)-(UStar+KStar)))
    
    #np.random.seed(t+1) # NOT needed!
    u = np.random.random(1)
    if u < alpha:
        x[:,t+1] = xStar.reshape(2,)
    else:
        x[:,t+1] = x[:,t]
        
    t += 1 # remember the last step

# Display
plt.scatter(x[0,:],x[1,:],c='black',label='Samples')
plt.plot(x[0,0:50],x[1,0:50],'ro-',markersize=1,linewidth=2,label="First 50 Samples")
# plt.xlim(xmin=-6,xmax=6)
# plt.ylim(ymin=-6,ymax=6)
plt.title("Hamiltonian Monte Carlo")
plt.legend(loc="upper left") # use labels for each plt.plot or plt.scatter, so the legend can show them directly
```
