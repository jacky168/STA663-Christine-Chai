Background for Exercises 1 and 2
----

The first 2 exercises are about using Newton's method to find the cube roots of unity - find $z$ such that $z^3 = 1$. From the fundamental theorem of algebra, we know there must be exactly 3 complex roots since this is a degree 3 polynomial.

We start with Euler's fabulosu equation
$$
e^{ix} = \cos x + i \sin x
$$

Raising $e^{ix}$ to the $n$th power where $n$ is an integer, we get from Euler's formula with $nx$ substituting for $x$
$$
(e^{ix})^n = e^{i(nx)} = \cos nx + i \sin nx
$$

Whenever $nx$ is an integer multiple of $2\pi$, we have
$$
\cos nx + i \sin nx = 1
$$

So
$$
e^{2\pi i \frac{k}{n}}
$$
is a root of 1 whenever $k/n = 0, 1, 2, \ldots$.

So the cube roots of unity are $1, e^{2\pi i/3}, e^{4\pi i/3}$. 

While we can do this analytically, the idea is to use Newton's method to find these roots, and in the process, disover some rather perplexing behavior of Newton's method.

**Exercise 1 (10 points)**. Netwon's method for functions of complex variables - stability and basins of attraction.

1. Write a function with the following function signature `newton(z, f, fprime, max_iter=100, tol=1e-6)` where
    - `z` is a starting value (a complex number e.g.  ` 3 + 4j`)
    - `f` is a function of `z`
    - `fprime` is the derivative of  `f`
The function will run until either max_iter is reached or the absolute value of the Newton step is less than tol. In either case, the function should return the number of iterations taken and the final value of `z` as a tuple (`i`, `z`). 

2. Define the function `f` and `fprime` that will result in Newton's method finding the cube roots of 1. Find 3 starting points that will give different roots, and print both the start and end points. 

```python
# your code here
def f(z):
    return (z**3 -1)

def fprime(z):
    """The derivative of f"""
    return (3*z**2)

def newton(z, f, fprime, max_iter=100, tol=1e-6):
    """Finding the cube roots of 1"""
    init = z # pass by value
    nextpt = z
    for i in range(max_iter):
        if np.abs(f(nextpt)) < tol:
            break
        # else (omitted)
        nextpt = nextpt - f(nextpt)/fprime(nextpt)
    return (i, nextpt)

# %precision 4
# np.set_printoptions(formatter={'float': '{: 0.3f}'.format})
print 'Using Newton\'s method to find the cube roots of 1:'
(iters1, nextpt1) = newton(3+4j,f,fprime)
print 'First starting point:   3+4j, root discovered: ({0.real:.4f}+{0.imag:.4f}j)'.format(nextpt1)
(iters2, nextpt2) = newton(-5+5j,f,fprime)
print 'Second starting point: -5+5j, root discovered: ({0.real:.4f}+{0.imag:.4f}j)'.format(nextpt2)
(iters3, nextpt3) = newton(-5-10j,f,fprime)
print 'Third starting point: -5-10j, root discovered: ({0.real:.4f}{0.imag:.4f}j)'.format(nextpt3)
```

**Exercise 2 (10 points)**. Write the following two plotting functions to see some (pretty) aspects of Newton's algorithm in the complex plane.

1. The first function `plot_newton_iters(f, fprime, n=200, extent=[-1,1,-1,1], cmap='hsv')` calculates and stores the number of iterations taken for convergence (or max_iter) for each point in a 2D array. The 2D array limits are given by `extent` - for example, when `extent = [-1,1,-1,1]` the corners of the plot are `(-i, -i), (1, -i), (1, i), (-1, i)`. There are `n` grid points in both the real and imaginary axes. The argument `cmap` specifies the color map to use - the suggested defaults are fine. Finally plot the image using `plt.imshow` - make sure the axis ticks are correctly scaled. Make a plot for the cube roots of 1.

2. The second function `plot_newton_basins(f, fprime, n=200, extent=[-1,1,-1,1], cmap='jet')` has the same arguments, but this time the grid stores the identity of the root that the starting point converged to. Make a plot for the cube roots of 1 - since there are 3 roots, there should be only 3 colors in the plot.

```python
# Part 1
def plot_newton_iters(f, fprime, n=200, extent=[-1,1,-1,1], cmap='hsv'):    
    xs = np.linspace(extent[0],extent[1],n)
    ys = np.linspace(extent[2],extent[3],n)
    X, Y = np.meshgrid(xs, ys)
    points = (np.vstack((X.ravel(), Y.ravel())))
    iters = np.array([newton(complex(points[0,i],points[1,i]),f,fprime)[0] for i in range(points.shape[1])]).reshape(n,n)
    
    plt.imshow(iters,cmap=cmap,extent=extent)
    plt.title('Number of iterations needed for convergence',y=1.05)
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.colorbar()
    pass

plot_newton_iters(f,fprime)
```

```python
# Part 2
def plot_newton_basins(f, fprime, n=200, extent=[-1,1,-1,1], cmap='jet'):
    xs = np.linspace(extent[0],extent[1],n)
    ys = np.linspace(extent[2],extent[3],n)
    X, Y = np.meshgrid(xs, ys)
    points = (np.vstack((X.ravel(), Y.ravel())))
    nodes = np.array([newton(complex(points[0,i],points[1,i]),f,fprime)[1] for i in range(points.shape[1])])

    # Classification: from nodes to roots
    node1 = 1; node2 = complex(-0.5,0.5*(3**0.5)); node3 = complex(-0.5,-0.5*(3**0.5))
    nodes_fac = np.zeros(len(nodes))
    
    for i in range(len(nodes)):
        if np.allclose(nodes[i],node1):
            nodes_fac[i] = 1
        elif np.allclose(nodes[i],node2):
            nodes_fac[i] = 2
        else:
        # np.allclose(nodes[i],node3):
            nodes_fac[i] = 3
    
    nodes_fac = nodes_fac.reshape(n,n)
    
    plt.imshow(nodes_fac,cmap=cmap,extent=extent)
    plt.title('The roots which each point converged to',y=1.05)
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    #plt.colors()
    pass

plot_newton_basins(f,fprime)
```

**Exercise 3 (10 points)**. Consider the following function on $\mathbb{R}^2$:

$$f(x_1,x_2) = -x_1x_2e^{-\frac{(x_1^2+x_2^2)}{2}}$$

```python
# Part 0: Setup
from sympy import symbols, hessian, Function, exp, Function, diff, solve, evalf
x, y = symbols('x y')
f = symbols('f', cls=Function)
f = -x*y*exp(-(x**2+y**2)/2.0)
print "f(x,y) =",f
```

```python
# Part 1: Gradient
f_grad = np.array([diff(f,x),diff(f,y)])
print "Gradient of f:\n", f_grad
```

```python
# Part 2: Hessian
f_H = hessian(f, (x, y))
print "Hessian of f:\n", f_H
```

```python
# Part 3: Critical points
f_crit = solve(f_grad,[x,y])
print "Critical points of f:\n",f_crit
```

```python
# Part 4-1: Characterize the critical points as max/min or neither
import scipy.linalg as la
f_H_crit = list()
for i in range(len(f_crit)):
    f_H_crit.append(f_H.subs([(x,f_crit[i][0]),(y,f_crit[i][1])]))
    print "Critical point",(i+1),":",f_crit[i]
    print "Hessian:",f_H_crit[i]
    print "Eigenvalues:",np.real_if_close(la.eig(f_H_crit[i])[0]),'\n'

print "Local min <=> positive definite <=> all eigenvalues positive"
print "Local max <=> negative definite <=> all eigenvalues negative"
print "Local min: (-1,-1), (1,1)"
print "Local max: (-1,1), (1,-1)"
print "Neither (saddle point): (0,0)"
```

```python
# Part 4-2: Find the minimum under the constraint
import scipy
from scipy import optimize as opt
# opt.minimize can only pass x = (x[0],x[1]) one object, not two!

def f2(x):
    return -x[0]*x[1]*exp(-(x[0]**2+x[1]**2)/2.0)

cons = ({'type': 'eq',
         'fun' : lambda x: np.array([2.0*x[0]+3.0*x[1]-5]),
         'jac' : lambda x: np.array([2.0,3.0])},
        {'type': 'ineq', # always "larger than"
         'fun' : lambda x: np.array([10-x[0]**2-x[1]**2])})

print "The constrained minimum is -0.3679, which occurs at the point (1,1)."
print opt.minimize(f2, (-5,-5), constraints=cons) # depends on the starting point
```

**Exercise 4 (20 points)**. Find the maximum likelihood function for a logistic regression. Load the file "train.csv" (located in this directory) into python.  This file is data from the survival data from the Titanic (from https://www.kaggle.com/).  You are to use a logistic regression to model survival as a function of gender, age and class (of travel). Find the maximum likelihood estimator of $\beta$ by numerical optimization using stochastic gradient descent as follows:

1. Stochastic gradient descent.  In this method, gradient descent is used essentially by fitting *one data point at a time*.  Recall the usual gradient descent step:

$$\beta_{i+1} = \beta_i - \nabla \ell(\beta_i)$$

where

$$\nabla \ell(\beta_i) = \sum\limits_{j=1}^n \nabla\ell(\beta_i,x_j,y_j)$$

and $\ell$ is the log-likelihood function.  All of the data is used to make the next step toward the optimal $\beta$.  In stochastic gradient descent, only one point at a time is used to determine the next $\beta$:

$$\beta_{i+1} = \beta_i - \alpha\nabla \ell(\beta_i,x_j,y_j)$$

where $\alpha$ is the step size.  For simplicity, we'll take a constant $\alpha=1$.

Implement the following stochastic gradient algorithm:
 
   a. Shuffle data points (i.e. randomly permute the order of the $(x_j,y_j)$
   
   b. Refine beta using the iterative formula above over each data point. 
   
   c. Repeat a and b until convergence is reached.
 
Apply the algorithm to the given data set to find the best-fit logistic regression coefficients.  Do not forget that your optimization should include a tolerance and a limit on the number of iterations.

```python
# Step 0: Load and pre-process the dataset
train = pd.read_csv('train.csv')
train = train.ix[:,['Survived','Pclass','Sex','Age']]
train = train.dropna()  # Some rows have NaNs => remove them

# Convert categorical variables into multiple 0/1 columns
# Categorical variables: Pclass (1/2/3) and Sex (male/female)
n = len(train.index) # 714

train['Pclass1'] = np.zeros(n)
train['Pclass2'] = np.zeros(n)
train['Pclass3'] = np.zeros(n)
train['SexM'] = np.zeros(n)
train['SexF'] = np.zeros(n)

for i in train.index:
    if train['Pclass'][i] == 1:
        train['Pclass1'][i] = 1
    elif train['Pclass'][i] == 2:
        train['Pclass2'][i] = 1
    else: # train['Pclass'][i] == 3
        train['Pclass3'][i] = 1
    
    if train['Sex'][i] == "male":
        train['SexM'][i] = 1
    else: # train['Sex'][i] == "female"
        train['SexF'][i] = 1

train = train.drop(['Pclass','Sex'], axis=1)        
        
train['Intercept'] = np.ones(n)

train = train[['Survived','Intercept','Age','Pclass1','Pclass2','SexM']]
# Re-order the variables of the dataset, and remove 'Pclass3' and 'SexF' to avoid collinearity

print train
```

```python
def logit_mle(beta, y, x):
    """Minus log likelihood function for logistic regression."""
    # print "x beta:",np.dot(x, beta).shape
    return -((-np.log(1 + np.exp(np.dot(x, beta)))) + (y*(np.dot(x, beta))))

    # return -((-np.log(1 + np.exp(x*beta))) + (y*x*beta))

def logit_gradient(beta, y, x):
    """Partial derivative of the logit_mle."""
    # temp = np.array()
    # print "gradient:",temp.shape
    return (np.reciprocal(1 + np.exp(-np.dot(x, beta))) - y)*x.T

def logit_optimize(data, f=logit_mle, fprime=logit_gradient, max_iter=100, tol=1e-10):
    """Stochastic gradient descent approach to find the beta.mle"""
    alpha = 0.01
    mle_init = 4.00
    for i in range(max_iter):
        #np.random.seed(21)
        indices = np.random.permutation(data.index)
        y = np.array(data.ix[indices,0])
        x = np.array(data.ix[indices,1:])
        #np.random.seed(28)
        beta_init = np.random.rand(x.shape[1],)
        beta_next = np.random.rand(x.shape[1],) # np.ones(x.shape[1],)
        
        for j in range(len(indices)): 
            beta_next = beta_next - alpha*fprime(beta_next,y[j],x[j,:])
            mle_next = f(beta_next,y[j],x[j,:])
        
        # Convergence criterion: MLE, not beta (different betas may obtain the same log likelihood)
        if abs(mle_next - mle_init) < tol:
            break

        beta_init = beta_next
        mle_init = mle_next
            
    return (i+1,beta_next,mle_next)

np.set_printoptions(formatter={'float': '{: 0.3f}'.format})
(iters, beta, mle) = logit_optimize(train)
print "Logistic regression:"
print iters,"iterations performed"
print "beta = [Intercept, Age, Pclass1, Pclass2, SexM]"
print "values",beta
print "The negative log likelihood is",mle
```
